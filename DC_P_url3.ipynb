### <font color = 'diamond'> To import packages

```python
import pandas as pd
from requests import get  # it allows to get the content (html) of webpage
from bs4 import BeautifulSoup as bs # store the content in beautiful in beautifulSoup object and sparse the code
```

```python
url = 'https://sn.coinafrique.com/categorie/appartements'
# get the html code of the page using the get function requests
res = get(url)
# store the html code in a beautifulsoup objet with a html parser (a parser allows to easily navigate through the html code)
soup = bs(res.content, 'html.parser' )
# get all containers that contains the informations of each motos and sooter
containers = soup.find_all('div', class_ = "col s6 m4 l3")
# length of containers
len(containers)
```

```python
# But we need to make sure the container we choose contain all the informations we need
container = containers[0]
```

```python
#scrape the following informations:
#ad type (vente ou location),number of rooms, price ,address,image_link
container_url = "https://sn.coinafrique.com/" + container.find('a')["href"]
# get the html code of the container
res_container = get(container_url)
soup_container = bs(res_container.content, "html.parser")
details = soup_container.find('h1',"title title-ad hide-on-large-and-down").text
price = "".join(soup_container.find('p',"price").text.strip().split()).replace('CFA','')
address = soup_container.find_all('span','valign-wrapper')[1].text
#get the apartment details
p_details = soup_container.find_all('div', class_="details-characteristics")[0]
j=p_details.find_all('span','qt')
number_of_rooms = j[0].text.strip() if len(j)>0 else None
img = soup_container.find('div',class_="swiper-slide slide-clickable")
style = img.get('style')
image_link=style.split('url(')[1].split(')')[0].strip('"')
```

```python
# scrape data from all the containers
data = []

for container in containers:
    container_url = "https://sn.coinafrique.com/" + container.find('a')["href"]
    res_container = get(container_url)
    soup_container = bs(res_container.content, "html.parser")
    try:
        details = soup_container.find('h1',"title title-ad hide-on-large-and-down").text
        price = "".join(soup_container.find('p',"price").text.strip().split()).replace('CFA','')
        address = soup_container.find_all('span','valign-wrapper')[1].text
        a_details = soup_container.find_all('div', class_="details-characteristics")[0]
        j=a_details.find_all('span','qt')
        number_of_rooms = j[0].text.strip() if len(j)>0 else None
        img = soup_container.find('div',class_="swiper-slide slide-clickable")
        style = img.get('style')
        image_link=style.split('url(')[1].split(')')[0].strip('"')

        dic = {
            "details": details,
            "price":price,
            "address":address,
            "a_details": p_details,
            "number_of_rooms":number_of_rooms,
            "image_link": image_link,
              }
        data.append(dic)
    except:
        pass
df = pd.DataFrame(data)
df.head()
```

### <font color = 'diamond'> Collect data over multiple pages

```python
n = int(input("How many pages do you want to scrape?" ))
```

```python
# create a empty dataframe df
df = pd.DataFrame()
# loop over pages indexes
for index in range(1, n):
  url = f'https://sn.coinafrique.com/categorie/appartements?page={index}'
  res = get(url)
  soup = bs(res.content, 'html.parser' )
  containers = soup.find_all('div', class_ = "col s6 m4 l3")

 # scrape data from all the containers
  data = []

for container in containers:
    container_url = "https://sn.coinafrique.com/" + container.find('a')["href"]
    res_container = get(container_url)
    soup_container = bs(res_container.content, "html.parser")
    try:
        details = soup_container.find('h1',"title title-ad hide-on-large-and-down").text
        price = "".join(soup_container.find('p',"price").text.strip().split()).replace('CFA','')
        address = soup_container.find_all('span','valign-wrapper')[1].text
        a_details = soup_container.find_all('div', class_="details-characteristics")[0]
        j=a_details.find_all('span','qt') # a_details stand for apartment details
        number_of_rooms = j[0].text.strip() if len(j)>0 else None
        img = soup_container.find('div',class_="swiper-slide slide-clickable")
        style = img.get('style')
        image_link=style.split('url(')[1].split(')')[0].strip('"')

        dic = {
            "details": details,
            "price":price,
            "address":address,
            "a_details": a_details,
            "number_of_rooms":number_of_rooms,
            "image_link": image_link,
              }
        data.append(dic)
    except:
        pass
DF = pd.DataFrame(data)
df = pd.concat([df, DF], axis = 0).reset_index(drop = True)
```

```python
df.price = pd.to_numeric(df.price, errors='coerce').astype('Int64')
df.number_of_rooms = pd.to_numeric(df.number_of_rooms, errors='coerce').astype('Int64')
df.dtypes
```

```python
#To resolve the missing values
median_price = df['price'].median()
df['price'] = df['price'].fillna(int(median_price)).astype('Int64')

median_number_of_rooms = df['number_of_rooms'].median()
df['number_of_rooms'] = df['number_of_rooms'].fillna(int(median_number_of_rooms)).astype('Int64')
```

```python
# check for null values
print(df.isnull().any().sum())
```

```python
import bs4

df = df.applymap(lambda x: str(x) if isinstance(x, bs4.element.Tag) else x)
```

### <font color = 'diamond'> To store data in a database

```python
import sqlite3
import pandas as pd
conn = sqlite3.connect('appartements1.db')
c = conn.cursor()
c.execute('''CREATE TABLE appartements1_table(details,price,address,a_details,number_of_rooms,image_link)''')
```

```python
df.to_sql(
    "appartement_table",
    conn,
    if_exists="append",
    index=False
)
conn.commit()
df_sql = pd.read_sql_query("SELECT * FROM appartement_table", conn)
conn.close()
df_sql.head()
```
